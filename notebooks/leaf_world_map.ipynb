# Food Leaf World Map

Interactive prototype for exploring where selected edible leaves are commonly used in cuisine.

The current database is provisional and seeded from curated LLM-assisted culinary intuition. Treat scores as starting hypotheses for review, not authoritative measurements.

## Setup

Create and activate the Conda environment from the project root with:

```bash
conda env create -f environment.yml
conda activate foodmap
jupyter lab notebooks/leaf_world_map.ipynb
```

````python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

SRC = PROJECT_ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

PROJECT_ROOT
````

````python
from leafmap import build_map_app, load_leaf_metadata, load_usage_data

usage = load_usage_data(PROJECT_ROOT / "data" / "leaf_usage_regions.csv")
metadata = load_leaf_metadata(PROJECT_ROOT / "data" / "leaf_metadata.csv")
````

## Data Coverage

````python
coverage = (
    usage.groupby(["leaf_id", "leaf_name"])
    .agg(rows=("admin1", "count"), countries=("country", "nunique"), avg_score=("usage_score", "mean"))
    .reset_index()
    .sort_values(["rows", "avg_score"], ascending=[False, False])
)
coverage
````

## Interactive Map

Select one or more leaves in the side panel. Colored bubbles show estimated regional usage. Black outlined circles mark regions where more than one selected leaf appears in the local database.

````python
build_map_app(usage, metadata)
````
